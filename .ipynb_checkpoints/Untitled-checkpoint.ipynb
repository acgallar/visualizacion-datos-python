# Bases de datos propinas

__Objeto a estudiar:__ 
* `tip`: Cantidad de propina. float

__Atributos de la base de datos__
* `sex`: Sexo del camarero/a. String
* `smoker`: zona fumadores. String
* `day`: Día de la semana. String
* `time`: Horario. Almuerzo/Cena String
* `size`: Tamaño de ma mesa. int
* `total_bill`: Valor total de la cuenta. float

```python
%matplotlib inline 

# Importamos Pandas. Para manipular tablas
import pandas as pd

# Matplotlib para gráficos
import matplotlib.pyplot as plt 

#Tipo de gráfico
plt.style.use('seaborn')

#Librería de herramientas estadísticas de visualización
import seaborn as sns

#Para regresión
import statsmodels.api as statm
import statsmodels.formula.api as frm
```

```python
# Vamos a importar nuestro csv
df = pd.read_csv('./tips.csv')
```

```python
df.describe()
```

## Esto usamos para describir las variables discretas

Calculamos frecuencia relativa

```python
df.sex.value_counts() / len(df)
```

```python
df.smoker.value_counts() / len(df)
```

```python
df.day.value_counts() / len(df)
```

```python
df.time.value_counts() / len(df)
```

## Vamos a generar una nueva variable Boolean para analizar variables String

```python
df.sex.replace(['Female','Male'],[1,0],inplace=True) 
#inplace reemplaza los datos en el mismo data frame, en caso
```

```python
df.smoker.replace(['No','Si'],[0,1],inplace=True) 
```

```python
df.time.replace(['Dinner','Lunch'],[0,1],inplace=True) 
```

```python
df.assign(weekend = df.day.map({'Sat': 1, 'Sun': 1, 'Thur': 0, 'Fri': 0}))
```

```python
df.total_bill.plot(kind = 'hist')
```

# Gráficos 

```python
#Tabla Cruzada
cross = pd.crosstab(df['sex'] , df['size'])
print(cross)
```

```python
cross.plot(kind = 'bar')
```

```python
cross2 = pd.crosstab(df['smoker'] , df['sex'])
cross2.plot(kind = 'bar')
```

## Creamos una matriz de correlaciones

```python
matrix_corr = df.corr()
matrix_corr
```

```python
sns.heatmap(matrix_corr, annot = True)
```

```python
sns.pairplot(df)
```

## Regresión

```python
model_1 = frm.ols('tip ~ total_bill', data=df)
model_1 = model_1.fit()
model_1.summary()
```
